```python
%pylab inline
#define default figure size
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

#def to convert cm to inches for plotting
def cm2inch(value):
    return value/2.54
```

```python
#Custom figure size in inches converted to cm using above def
figure(figsize=(cm2inch(15.0),cm2inch(12.0)))

x = linspace(-5,5,100);
plot(x,x**2,'--,g',label='line 1');
title('Test')
legend();
```
