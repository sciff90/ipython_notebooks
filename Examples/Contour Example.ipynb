```python
%pylab inline
#default figure size
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
delta = 0.025;
x = arange(-3.0,3.0,delta);
y = arange(-2.0,2.0,delta);

#Meshgrid maps x,y to 2d grid defined by X,Y
X, Y = meshgrid(x, y)


Z1 = mlab.bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
Z2 = mlab.bivariate_normal(X, Y, 1.5, 0.5, 1, 1)
# difference of Gaussians
Z = 10.0 * (Z2 - Z1);

#Contour Lines
figure()
CS = contour(X,Y,Z,10);
clabel(CS,fontsize = 10);
```

```python
#Filled Contours
figure()
CS = contourf(X,Y,Z,100);
```
